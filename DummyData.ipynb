# **exam_data**

```python
def generate_exam_data():
    # Fetch a random student record to ensure the student_email exists
    student_response = supabase.table('student_records').select('email', 'name', 'class', 'section', 'institute_id').execute()
    if not student_response.data:
        raise ValueError("No student records found. Please ensure the student_records table has data.")
    student = student_response.data[0]

    # Fetch a random exam_name record to ensure fk_exam_id exists
    exam_response = supabase.table('exam_names').select('id').limit(1).execute()
    if not exam_response.data:
        raise ValueError("No exam names found. Please ensure the exam_names table has data.")
    exam = exam_response.data[0]

    # Fetch a random institute record to ensure institute_id exists
    institute_response = supabase.table('institute_records').select('id').limit(1).execute()
    if not institute_response.data:
        raise ValueError("No institute records found. Please ensure the institute_records table has data.")
    institute = institute_response.data[0]

    return {
        "student_email": student['email'],
        "subject": random.choice(['Mathematics', 'Science', 'English', 'History', 'Geography']),
        "exam_name": random.choice(['Midterm', 'Finals', 'Unit Test']),
        "score": round(random.uniform(35, 100), 2),
        "max_score": 100,
        "class": student['class'],
        "section": student['section'],
        "institute_id": institute['id'],
        "fk_exam_id": exam['id'],
        "student_name": student['name']
    }
```

```python
exam_data = [generate_exam_data() for _ in range(50)]
```
